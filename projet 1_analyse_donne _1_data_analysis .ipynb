```python
#Ajoute 3 lignes vides à la fin d'un DataFrame : 
import pandas as pd
import numpy as np
empty_rew = pd.DataFrame([ {
     "gender" : np.nan,
   "race/ethnicity" : np.nan,
   "parental level of education" : np.nan,
   "lunch" : np.nan,
   "test preparation course" : np.nan,
   "math score" : np.nan,
   "reading score" : np.nan,
   "writing score" : np.nan
} for _ in range(3)])
df =pd.concat([df,empty_rew],ignore_index =True )
df.tail()
```

```python
nb =df.isnull().sum()
print(nb)
df[df.isnull().any(axis =1)]
```

```python
#Remplacer les valeurs manquantes dans la colonne 'math score' par la valeur moyenne de cette colonne
df["math score"]=df["math score"].fillna(df["math score"].mean())
df.tail()
```

```python
#Remplacer les valeurs manquantes dans la colonne 'gender' par le texte 'Unknown'
df["gender"]=df["gender"].fillna("Unknown")
df.tail()
```

```python
#Remplacer les valeurs manquantes (NaN) par la dernière valeur non manquante précédente 
df =df.ffill()
df.tail()
```

```python
#Remplace les valeurs manquantes (NaN) par la prochaine valeur non manquante suivante
df =df.bfill()
df.tail()
```

```python
#Calculer la médiane de la colonne et remplacer les NaN en une seule étape 
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy ="median")
df[["reading score"]] = imputer.fit_transform(df[["reading score"]])
df.tail(5)
```

```python
# Afficher uniquement les lignes où la colonne 'model' contient des valeurs manquantes
df[df['lunch'].isnull()]
```

```python
#Remplace les valeurs manquantes dans la colonne 'reading score' par 50, mais seulement pour les etudient de la gender female	
df.loc[(df["reading score"].isnull())&(df["gender"]=="female"),'reading score']=50
df.tail(5)
```

```python
#Remplacer toutes les valeurs manquantes dans la colonne 'lunch' par la valeur 'Standard' 
#Remplacer les valeurs manquantes dans la colonne 'writing score' par la valeur médiane de cette colonne 
df.loc[df["lunch"].isnull(),'lunch']="Standard"
df.loc[df["writing score"].isnull(),'writing score']=df["writing score"].median()
df.tail(5)
```

```python
#Calculer l'intervalle interquartile (IQR) pour la colonne 'reading score', 
#Identifier les valeurs aberrantes (outliers) dans la colonne 'reading score' 
Q1=df ['reading score'].quantile(0.25)
Q3=df ['reading score'].quantile(0.75)
IQR = Q3 - Q1
outliers =df [(df ['reading score']<Q1-1.5*IQR) | (df ['reading score']>Q3+1.5*IQR)]
outliers
```

```python
#Utiliser le Z-score pour détecter les valeurs aberrantes dans la colonne 'math score'
from scipy.stats import zscore
df['zscore'] =zscore(df['math score'])
df[df['zscore'].abs()>3]
```

```python
import pandas as pd

# Load your StudentsPerformance.csv file
df_StudentsPerformance = pd.read_csv('StudentsPerformance.csv')

# Split and concatenate
df1 = df_StudentsPerformance.iloc[:5]
df2 = df_StudentsPerformance.iloc[5:10]

# Row-wise concat (stack rows)
concat_rows = pd.concat([df1, df2], axis=0)
print("Row-wise concatenation:")
print(concat_rows.head(10))

# Column-wise concat (side by side)
concat_cols = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
print("\nColumn-wise concatenation:")
print(concat_cols.head())
```
